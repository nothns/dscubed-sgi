```python
import pandas as pd
import numpy as np
from collections import defaultdict
import re
```

```python
df = pd.read_csv('data.csv')
```

```python
df.columns
```

```python
df.rename(columns={'Subject 1 (In code format, i.e. MAST10007)': 'Subject 1'}, inplace=True)
```

```python
for column in df.columns[-4:]:
    df[column] = df[column].str.upper()
```

```python
df1 = df[['Full Name', 'Student Email', 'Subject 1', 'Subject 2', 'Subject 3', 'Subject 4']]
```

```python
# for subjects 1-4 columns, if the value is not made up of FOUR letters followed by 5 numbers using regular expression, then make NA

for column in df1.columns[-4:]:
    df1[column] = df1[column].apply(lambda x: x if re.match(r'[A-Z]{4}[0-9]{5}', str(x)) else np.nan)

df1
```

```python
# change the first value of the 7th row to Jean

df1.iloc[7, 0] = 'Jean'
```

```python
df1
```

```python
subjects_dict = defaultdict(list)

# Iterate over each row in DataFrame
for index, row in df1.iterrows():
    email = row['Student Email']

    # Iterate over each subject column
    for subject_col in df1.columns[2:]:
        subjects_dict[row[subject_col]].append(email)

del subjects_dict[np.nan]  # Delete the key for NaN values
```

```python
subjects_dict
```

```python
# Create a DataFrame from the dictionary
df_new = pd.DataFrame(list(subjects_dict.items()), columns=['Subject', 'Names'])

# Add a count column by applying len function to 'Names' column
df_new['Count'] = df_new['Names'].apply(len)
df_new
```

```python
grouped_dict = defaultdict(list)

# Loop over each subject in the old dictionary
for subject, students in subjects_dict.items():
    # Special handling for groups of 5 or less
    if len(students) <= 5:
        if len(students) >= 3:  # If the group has 3, 4 or 5 students, keep them as a single group
            grouped_dict[subject].append(students)
        else:  # If the group has less than 3 students, mark it as 'N/A'
            grouped_dict[subject].append('N/A')
    else:  # If there are more than 5 students
        while len(students) >= 3:
            # Create groups of 3 as base unit
            if len(students) % 3 == 0 or len(students) % 3 == 1: # When count of students is a multiple of 3 or remainder 1
                group = students[:3]
                students = students[3:]
            else: # When the remainder is 2, make a group of 4 to prevent a group of 2 in the next step
                group = students[:4]
                students = students[4:]
            grouped_dict[subject].append(group)
        if students:  # If there are 1 or 2 students left (which should not happen), mark them as 'N/A'
            grouped_dict[subject].append('N/A')

grouped_dict
```

```python
# Create a new dictionary to hold the email groups
email_dict = {email: defaultdict(list) for email in df['Student Email'].unique()}

for subject, groups in grouped_dict.items():
    for group in groups:
        if group == 'N/A':
            continue
        for student_email in group:
            # Create a list of group member emails excluding the current student
            email_dict[student_email][subject] = group

# Now, email_dict is a dictionary that maps each student to a dictionary.
# Each inner dictionary maps subjects to lists of lists (i.e., groups of group member emails)
email_dict
```

```python
# Create a list to hold the data
data = []

# Go through each student in email_dict
for student_email, groups in email_dict.items():
    # Count the number of subjects the student has been matched to
    num_subjects = len(groups)
    # Append the data to the list
    data.append([student_email, num_subjects])

# Create a DataFrame from the data
df_check = pd.DataFrame(data, columns=['Student Email', 'Number of Subjects'])

# Display the DataFrame
df_check
```

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

def send_email(student_email, msg_content):
    from_alias = "Study Groups Initiative"  # replace with your name
    from_address = "nathan.luo@dscubed.org.au"  # replace with your email
    password = "nmrocrhtfymydzsz"  # replace with your password

    # Setup the MIME
    message = MIMEMultipart()
    message['From'] = f"{from_alias} <{from_address}>"
    message['To'] = student_email
    message['Subject'] = "Your Study Groups"   # Subject of the email

    # Attach the message to the MIME object
    message.attach(MIMEText(msg_content, 'plain'))

    # Use SMTP to send the email
    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)  # Use Gmail's SMTP server
        server.starttls()  # Start TLS encryption
        server.login(from_address, password)
        text = message.as_string()
        server.sendmail(from_address, student_email, text)
        server.quit()
        print(f"Email sent to {student_email}")
    except Exception as e:
        print(f"Failed to send email to {student_email} due to {str(e)}")
```

```python
def compose_email(df, student_email, student_groups):
    # Look up the full name from the DataFrame
    full_name = df.loc[df['Student Email'] == student_email, 'Full Name'].values[0]
    # Extract the first name
    first_name = full_name.split()[0]

    # Compose the email
    email_body = f"""
Dear {first_name},

I hope this message finds you well.

Thank you for your enthusiasm and for signing up to be a part of our Study Groups Initiative. We're thrilled to have you join us and appreciate your commitment to collaborative learning.

Below you will find the details for your assigned study groups:
"""
    # Add each group and subject to the email body
    for subject in student_groups[student_email].keys():
        email_body += f"\nSubject: {subject}\nMembers: "
        for member in student_groups[student_email][subject][:-1]:
            email_body += f"{member}, "
        email_body += f"{student_groups[student_email][subject][-1]}\n"

    email_body += """
Please note, if a subject you have chosen does not appear above, it is because there was limited interest in that subject. We apologize for any inconvenience this may have caused.

We encourage you to reach out to your fellow group members by sending them an email and perhaps organizing a group chat on social media. This can help facilitate communication and planning for your study sessions.

If you have any questions or need further assistance, please don't hesitate to reach out. We're here to help!

Thank you again for your active participation in the Study Groups Initiative. We're excited to see how this collaboration enhances your learning experience.

Best regards,

Nathan
"""

    return email_body

# Test the function
student_email = 'jiacheng7@student.unimelb.edu.au'
student_groups = {
    'MAST10007': ['jiacheng7@student.unimelb.edu.au', 'student3@example.com'],
    'MAST10008': ['jiacheng7@student.unimelb.edu.au', 'student5@example.com']
}

for email in df1['Student Email']:
    email_body = compose_email(df1, email, email_dict)
    print(email_body)
```

```python
# Using the send_email function

for student_email, subjects in email_dict.items():
    if len(subjects) == 0:
        print("No subjects found for: ", student_email)
        continue
    msg_content = compose_email(df1, student_email, email_dict)
    send_email(student_email, msg_content)
```
